```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore') 
```

```python
cctv = pd.read_csv('서울시CCTV설치운영현황(자치구)_년도별_210731기준.csv',
                    skiprows=1, encoding='euc-kr')
cctv.head()
```

```python
cctv.drop(index=[0], inplace=True)
cctv.head()
```

```python
cctv.columns[1:]
```

```python
for column in cctv.columns[1:]:
    cctv[column] = cctv[column].str.replace(',','').astype(int)
cctv.head()
```

```python
cctv['최근증가율'] = cctv.iloc[:,1:8].sum(axis=1) / cctv.iloc[:,8:11].sum(axis=1) * 100
cctv.head()
```

```python
cctv2 = cctv[['구분','총계','최근증가율']]
cctv2.columns = ['구별','CCTV댓수','최근증가율']
cctv2.head()
```

```python
cctv_result = cctv2.set_index('구별', inplace=True)
cctv_result.head()
```

```python
pop = pd.read_csv('서울시주민등록인구(구별)통계.txt', sep='\t', encoding='utf-8',header=0)
pop.head()
```

```python
pop.drop(index=[0,1,2], inplace=True)
pop.head()
```

```python
pop.columns[2:]
```

```python
pop2 = pop[['자치구', '인구', '인구.6', '65세이상고령자']]
pop2.columns=['구별','인구수','외국인','고령자']
pop2.head()
```

```python
for column in pop2.columns[1:]:
    pop2[column] = pop2[column].apply(lambda x: int(x.replace(',','')))
pop2.head()
```

```python
pop2['외국인비율'] = pop2['외국인'] / pop2['인구수'] * 100
pop2['고령자비율'] = pop2['고령자'] / pop2['인구수'] * 100
pop2.head()
```

```python
pop_result = pop.set_index('구별', inplace=True)
pop_result.head()
```
