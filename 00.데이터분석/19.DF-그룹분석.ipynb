# DataFrame - 그룹 분석

```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
# 품종별, sepal_length의 평균
# 그냥 시리즈로 결과값 나옴
iris.groupby('species')['sepal_length'].mean()
```

```python
# 리스트 넣어서 표현해주면 데이터 프레임으로 변함
# 그룹핑 후 열 선택
iris.groupby('species')[['sepal_length']].mean()
```

```python
# 필요한 열 먼저 선택하고 그룹 분석
iris[['sepal_length','species']].groupby('species').mean()
```

```python
# 모든 피처에 대한 품종별 평균
iris.groupby('species').mean()
```

```python
# 모든 피처에 대한 품종별 평균, 표준편차
res = iris.groupby('species').agg(['mean','std'])
res
```

```python
# res는 컬럼 인덱스가 두 개
# 
res['sepal_length']['mean']['setosa']
```

- tips data

```python
tips = sns.load_dataset('tips')
tips.head()
```

- Guessing
   - 성별(남자), 흡연유무(흡연자), 점심/저녁(저녁)

```python
tips['tip_pct'] = (tips.tip / tips.total_bill * 100).round(2)
tips.head()
```

```python
# 성별에 따른 tip_pct의 평균
tips.groupby('sex')[['tip_pct']].mean()
```

```python
# 흡연 유무, 시간에 따른 tip_pct의 평균
tips.groupby(['smoker', 'time'])[['tip_pct']].mean()
```
