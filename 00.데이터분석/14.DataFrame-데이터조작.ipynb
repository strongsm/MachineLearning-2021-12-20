- apply 메소드

```python
df3 = pd.DataFrame({
     'A': [1, 3, 4, 3, 4],
     'B': [2, 3, 1, 2, 3],
     'C': [1, 5, 2, 4, 4]
})
df3
```

```python
# 선호하지 않는 방법
def func1(x):
    return x.max() - x.min()
```

```python
df3.apply(func1)
```

```python
# 인수 x
df3.apply(lambda x: x.max() - x.min())
```

```python
# 행 단위로 하고자 할 경우에는 axis=1 option을 줘야 함 
df3.apply(lambda x: x.max() - x.min(), axis=1)
```

```python
# 각 행별로 최대값에서 최소값을 뺀 결과를 '편차'라는 이름의 열로 만든다.
df
```

```python
df['편차'] = df.apply(lambda x: x.max() - x.min(), axis=1)
df
```

```python
# 평균이 70점 이상이면 PASS, 아니면 FAIL인 '결과'열을 추가할 것
df['결과'] = df.평균.apply(lambda x: 'PASS' if x >= 70 else 'FAIL')
df
```

```python
# 100점답
# 전체에 적용, axis=1
df['결과2'] = df.apply(lambda x: '합격' if x.평균 >= 70 else '불합격', axis=1)
df
```

```python
# 기초
def func2(x):
    if x.평균 >= 70:
        return '합격'
    else:
        return '불합격'
```

```python
# 1000 단위 구분기호 없애기
# np.random.randint(100,1000,1) 값이 array로 나오니까 [0] 첫번째 인덱스를 쓰겠다
price_list = []
for i in range(12):
    price = np.random.randint(100,1000,1)[0]*100
    price_list.append(f'{price:,d}')
price_list
```
